```python
import cv2               #OpenCV-Python is a library of Python bindings designed to solve computer vision problems.              


def generate_dataset(img,id,img_id):              # creating a new method to generate dataset and uising image , id ,img_id as arguments.....
                                                  # ..here "id" is used to distinguish a person instead of names , "img_id" is the id of the images.. 
                                                  #.. captured for dataset ,more the images are in dataset , better will be the recognition.
            
    cv2.imwrite("C:\\Users\\KIIT\\Downloads\\imageprocessing\\project\\data/user."+str(id)+"."+str(img_id)+".jpg",img) 
                                                  #passing the directory and concatinating the ids with image , and the image file itself.


def draw_boundary(img,classifier,scaleFactor,minNeighbors,color,text,clf):
    
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)                             #converting RGB img to gray scale img
    
    features= classifier.detectMultiScale(gray_img,scaleFactor,minNeighbors)  #here, through classifier we are using detectMulticale to detect all...
                                                                              #... the possible faces in the image.
                                                                              #Sacle factor-Basically the scale factor is used to create your scale..
                                                                              #..pyramid
                                                                              # minNeighbors- this parameter will affect the quality of the detected
                                                                              #..faces
                                                                        
    
    coords=[]                     # creating a list to hold the coordinates parameters
    
    for (x,y,w,h) in features:    # x- x axis , y-y axis , w - width , h- height
        
        cv2.rectangle(img,(x,y),(x+w,y+h),color,2)   # to draw rectangle around the face detected.
        id, _ = clf.predict(gray_img[y:y+h,x:x+w])   # using classifier "clf"  to predict the face over grayscale and making the ROI to our face only
       
        if id==1:        # if id ==1 then it will predict abhishek as my images were saved with user 1.01.jpg
            
            cv2.putText(img,"ABHISHEK",(x,y-4),cv2.FONT_HERSHEY_SIMPLEX,0.8,color,1,cv2.LINE_AA)   # to put text over the image
            
        coords =[x,y,w,h]         #updating the coordinates
        
    return coords

def recognize(img,clf,faceCascade):                                            #updating the coordinates
        .
    color={"blue":(255,0,0),"red":(0,0,255),"green":(0,255,0)}
      
    coords= draw_boundary(img,faceCascade,1.1,10,color["blue"],"Face",clf)
    return img





def detect(img,faceCascade,eyeCascade,noseCascade,img_id):      # declaring a function to detect face by taking  args i.e 
                                                        #...image and classifier of frontal face,eyes,nose and img_id 
                                                          
        
    color={"blue":(255,0,0),"red":(0,0,255),"green":(0,255,0)}                # declaring color as a dictionary in RGB Code.
    coords =draw_boundary(img,faceCascade,1.1,10,color['blue']," Face ") # passing arguments to detect faces in image and draw boundary..
                                                                              # across it. 
    
    if len(coords)==4:                                                      # if face detected it will all the four coordinates
        
        roi_img= img[coords[1]:coords[1]+ coords[3],coords[0]:coords[0]+coords[2]]   #creating region of interest for detecting only face ,nose and..
                                                                                     #...eyes 
        
        user_id=1          # setting user_id for any random first user as 1
        generate_dataset(roi_img,user_id,img_id)     # calling method to generate dataset
        

        #coords=draw_boundary(roi_img,eyeCascade,1.1,14,color['red']," Eyes ")  #  
        #coords=draw_boundary(roi_img,noseCascade,1.1,5,color['green']," nose ")   # passing arguments to detect nose in image and draw boundary..
                                                                              #... across it. 
        
    return img      
    

faceCascade=cv2.CascadeClassifier('C:\\Users\\KIIT\\Downloads\\imageprocessing\\project\\haarcascade_frontalface_default.xml') 
                                   # Loading haarcascade_frontalface_default.xml classifier file to faceCascade to detect faces.                                       
eyeCascade=cv2.CascadeClassifier('C:\\Users\\KIIT\\Downloads\\imageprocessing\\project\\haarcascade_eye.xml')
                                  # Loading haarcascade_eye.xml classifier file to eyeCascade to detect eyes.  
noseCascade=cv2.CascadeClassifier('C:\\Users\\KIIT\\Downloads\\imageprocessing\\project\\Nariz.xml')    
                                 # Loading Nariz.xml classifier file to noseCascade to detect nose.  
    
    

clf = cv2.face.LBPHFaceRecognizer_create()                                            # Creating clf 
clf.read("C:\\Users\\KIIT\\Downloads\\imageprocessing\\project\\classifier.yml")      # Loading our Classifier file .
    
    
video_capture = cv2.VideoCapture(0)   # it will grab our default camera.

img_id =0                            #initialising with 0.

while True:
    
    _,img=video_capture.read()       
                             #this video_capture. read() grabs the frame and it returns back tuple of image files continuously and keeps  updating.
     
    #img = detect(img,faceCascade,eyeCascade,noseCascade,img_id) 
                                               # calling detect function to detect faces and passing image and facecascade file as arguments.

    img = recognize(img,clf,faceCascade)     #  Calling Method to recognize face.  
        
    cv2.imshow("face detection",img)           #shows the image and face detection is the frame name .
    
    img_id +=1                                 #incrementing img_id by 1 after capturing each frame .
    
    if cv2.waitKey(1) & 0xFF == ord('q'):      #press q to exit the frame. and waitkey()is for delaying so that human can see the video.
        break

video_capture.release()                     #stops capturing
cv2.destroyAllWindows()

```
